```python
data = pd.read_csv('delivery_time.csv')
```

```python
data.head()
```

```python
data.describe()
```

```python
features= data[['Sorting Time']]
```

```python
features
```

```python
target = data[['Delivery Time']]
```

```python
target
```

```python
x_train,x_test,y_train,y_test = train_test_split(features,target,train_size=0.75,random_state=20)
```

```python
linreg=LinearRegression()
```

```python
linreg.fit(x_train,y_train)
```

```python
y_pred= linreg.predict(x_test)
```

```python
y_pred
```

```python
linreg.intercept_
```

```python
linreg.coef_
```

```python
linreg.score(x_test,y_test)
```

```python
import seaborn as sns
```

```python
sns.regplot(data,x=y_test,y=y_pred,scatter_kws={'color':'r'},line_kws={'color':'purple'})
```
